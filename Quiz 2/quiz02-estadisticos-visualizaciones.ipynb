### Actividad 1

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
data = pd.read_csv("customer_data.csv")
```

```python
data.head()
```

Technical_Issues: CUANTOS problemas han tenido en su vida como cliente

NPS: Net Promodet Score, del 1-10 que tanto te gusta el producto

CAC = Costo de adquicision de clientes

Expected_LTV_in_Two_Years = Cuanto espero que te de de valor el lciente en 2 años

Expected_Lifespan_Months: cuanto creo que durara conmigo el cliente en meses

Les cobor por mes  mis clientes 10.80

* Buscar la relcion entre NPS, LOS SEGMENTOS DE clientes o tipo de cliente y Lifesan


```python
corr = data.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')

plt.title('Mapa de Calor de la Correlación de Características Numéricas')

plt.show()
```

```python
data.groupby('NPS_Score')['Expected_Lifespan_Months'].mean().plot(kind='line', marker='o', color='purple')

plt.title('Promedio de expectativa de clientes por califiacion')

plt.xlabel('Meses')

plt.ylabel('Que tanto les gusta')

plt.show()
```

Vemos claramente que mientras mejor valoracion o mejor experiencia haya tenido un cliente durante el primer mes tendra mas expectativa de estar mas tiempo en la empresa. Asi que te recomiendo hacer que el primer mes del cliente sea lo que espera para que este mas tiempo y asi te generen mas.

```python
data.groupby('NPS_Score')['Technical_Issues'].mean().plot(kind='bar', color='skyblue')

plt.title('Cuantos errores tuvieron en promedio los que alificaron con 1,2,3,..,10')

plt.xlabel('Calificacion')

plt.ylabel("Error")

plt.show()
```

```python
data.groupby('Segment')['Expected_Lifespan_Months'].mean().plot(kind='bar', color='skyblue')

plt.title('En promedio cuanta expectativa de vida tiene cada tipo de cliente')

plt.xlabel('Tipo de cliente')

plt.ylabel("Promedio de tiempo en la empresa")

plt.show()
```

```python
data.groupby('Segment')['Expected_LTV_in_Two_Years'].mean().plot(kind='bar', color='skyblue')

plt.title('En promedio cuanto espero que me genere cada tipo de cliente')

plt.xlabel('Tipo de cliente')

plt.ylabel("Promedio de tiempo en la empresa")

plt.show()
```

```python
data.groupby('Segment')['NPS_Score'].mean().plot(kind='bar', color='skyblue')

plt.title('En promedio que tipo de cliente se sintio mas satisfecho')

plt.xlabel('Tipo de cliente')

plt.ylabel("Que clientes son mas satisfechos")

plt.show()
```

## Conclusion y recomendacion final
## Respuesta final

Como puedes observar, las gráficas son claras: deben centrarse en sus clientes "Young Professionals". 

Si yo fuera Floating Pandas, me enfocaría en estos clientes, ya que son los que, en promedio, tienen un mayor tiempo de vida, son los que mejor calificación dejan en el primer mes y, sobre todo, son los que más ganancia generan, que es lo que esta empresa busca.

Por otro lado, como vimos en las gráficas, sucede lo opuesto con los clientes "Senior".

Se podrían implementar sugerencias específicas para estos clientes, y si en el futuro se observan mejoras en el LTV (Lifetime Value), recomendaría seguir con ellos; de lo contrario, sería mejor buscar atraer a otros tipos de clientes.

```python
data.groupby('Segment')['CAC'].mean().plot(kind='bar', color='skyblue')

plt.title('En promedio que tipo de cliente te cuesta mas de traer')

plt.xlabel('Tipo de cliente')

plt.ylabel("CAC")

plt.show()
```

Interesante observar que en promedio se gasta lo mismo por atraer un tipo de cliente

```python
data["CAC_LTV"] = data["Expected_LTV_in_Two_Years"]/data["CAC"]
#Ratio o retorno de intersion
```

```python
#Promedio de retorno de inversion por segmento
data.groupby('Segment')['CAC_LTV'].mean()
```

```python
data_final = data.groupby(['Acquisition_Month',"Segment"])['CAC_LTV'].mean().reset_index()
data_final.head()
```

```python
sns.lineplot(x='Acquisition_Month', y='CAC_LTV', data=data_final, hue = "Segment" )

plt.title('Promedio del bench_mark en el tiempo por segmento')
plt.xticks(rotation=50) 
plt.show()
```

En el tiempo que benchmark o ratio tuvo cada grupo
